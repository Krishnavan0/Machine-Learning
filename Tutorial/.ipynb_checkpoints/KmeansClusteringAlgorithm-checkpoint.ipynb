```python
# Problem Statement
'''A Mall wants to understand its customer base better to improve its marketing strategy. The goal is to cluster customers into
groups based on their annual income and spending score, so the marketing team can tailor their campaigns to each cluster's 
preferences and behaviors.'''
```

```python
# step no 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# generate data set
np.random.seed(42)
data = {'Annual Income': np.random.randint(30000,100000,100),
        'Spending Score': np.random.randint(1,100,100)}

data = pd.DataFrame(data)

X = data.values
kmeans = KMeans(n_clusters=3, random_state=42)
data['Cluster'] = kmeans.fit_predict(X)

data.head()
```

```python
plt.scatter(data['Annual Income'],data['Spending Score'], c=data['Cluster'], cmap='rainbow')
plt.title('KMeans Customer Data - Annual income vs spending score')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

```python
# user input
user_input = {'Annual Income':[75000], 'Spending Score':[70],}
user_data = pd.DataFrame(user_input)
user_cluster = kmeans.predict(user_data)
print(f"The user belongs to Cluster {user_cluster[0]}")
```
